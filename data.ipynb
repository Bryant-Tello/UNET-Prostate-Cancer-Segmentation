```python
import pandas  as pd
import numpy   as np
import nibabel as nib
import matplotlib.pyplot as plt
import os 
from tqdm import tqdm 
from ipywidgets import interact, interactive, IntSlider, ToggleButtons
import cv2
```

```python
def read_nii(filepath):
    '''
    Reads .nii file and returns pixel array
    '''
    ct_scan = nib.load(filepath)
    array   = ct_scan.get_fdata()
    array   = np.rot90(np.array(array))
    return(array)
```

```python
def one_list(lista):
  img_list=[]
  for sub_lista in lista:
    for elemento in sub_lista:
      img_list.append(elemento)
  return img_list
```

```python
def read_image_nii(img, mask):
  data_img=[] 
  data_mask=[] 
  for image in zip(mask,img):
    mask = read_nii(data_path+image[0])
    img = read_nii(data_path+image[1])
    data_mask.append(mask)
    data_img.append(img)
  return data_img, data_mask
```

```python
@interact
def explore_3dimage(layer = (0,383),patient=(0,92)  , view = [ 'coronal','axial' , 'sagital'] ): 

    array_view_image = data_img_train[patient]
    array_view_mask = data_mask_train[patient]

    if view == 'axial':
      array_view_image = array_view_image[:, : ,layer]
      array_view_mask = array_view_mask[:, : ,layer]
    elif view == 'coronal':
      array_view_image = array_view_image[:,layer, :]
      array_view_mask = array_view_mask[:, layer, :]
    elif view == 'sagittal':
      array_view_image = array_view_image[layer, :,:]
      array_view_mask = array_view_mask[layer, :,:]
    else:
      print("Error")


    fig, (ax1, ax2) = plt.subplots(1,2, figsize = (20, 16))

    ax1.imshow(array_view_image)
    ax1.set_title('Image')

    ax2.imshow(array_view_mask)
    ax2.set_title('Mask')
```

```python
def normalizeImageIntensityRange(img):
    HOUNSFIELD_MIN = np.min(img)
    HOUNSFIELD_MAX =np.max(img)
    HOUNSFIELD_RANGE = HOUNSFIELD_MAX - HOUNSFIELD_MIN
    img[img < HOUNSFIELD_MIN] = HOUNSFIELD_MIN
    img[img > HOUNSFIELD_MAX] = HOUNSFIELD_MAX
    return (img - HOUNSFIELD_MIN) / HOUNSFIELD_RANGE
```

```python
dataOutputPath_Train = '/content/drive/MyDrive/Colab Notebooks/data/Train/'
img_Output_train = os.path.join(dataOutputPath_Train, 'Img/')
mask_Output_train = os.path.join(dataOutputPath_Train, 'Mask/')

dataOutputPath_test = '/content/drive/MyDrive/Colab Notebooks/data/Test/'
img_Output_test = os.path.join(dataOutputPath_test, 'Img/')
mask_Output_test= os.path.join(dataOutputPath_test, 'Mask/')
```

```python
def readImageVolume(img, normalize=False):
    if normalize:
        return normalizeImageIntensityRange(img)
    else:
        return img
```

```python
# Save volume slice to file
def saveSlice(img, fname, path,mask):
    if mask:
      img = (np.arange(3) == img[...,None]).astype(np.float32) 
    else:
      img = np.uint8(img * 255)
    fout = os.path.join(path, f'{fname}.png')
    cv2.imwrite(fout, img)
    print(f'[+] Slice saved: {fout}', end='\r')
    
```

```python
# Slice image in all directions and save
def sliceAndSaveVolumeImage(vol, fname, path,mask):
    (dimx, dimy, dimz) = vol.shape
    print(dimx, dimy, dimz)
    cnt = 0          
    cnt += dimz
    print('Slicing Z: '+str(dimz))
    for i in range(dimz):
      saveSlice(vol[:,:,i], fname+'-slice'+str(i), path,mask)
    return cnt
```

```python
# Read and process image volumes
count=0

for index in range(len(data_img_train)):
    img = readImageVolume(data_img_train[index], True)
    print('image '+str(index), img.shape, np.sum(img.shape), np.min(img), np.max(img))
    numOfSlices = sliceAndSaveVolumeImage(img, 'prostate'+str(index), img_Output_train,False)
    print(f'\n{+(index)}, {numOfSlices} slices created \n')
    print('*******************')
    mask = readImageVolume(data_mask_train[index], False)
    print('image '+str(index), mask.shape, np.sum(mask.shape), np.min(mask), np.max(mask))
    numOfSlices = sliceAndSaveVolumeImage(mask, 'prostate'+str(index), mask_Output_train,True)
    print(f'\n{+(index)}, {numOfSlices} slices created \n')
    print('***********************************')
    count = numOfSlices + count
print('')    
print(count)
```
